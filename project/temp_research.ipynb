```python
import torch
from tuned_lens.nn.lenses import TunedLens, LogitLens
from transformers import AutoModelForCausalLM, AutoTokenizer
from tuned_lens.plotting import PredictionTrajectory
from tempfile import TemporaryDirectory
from transformers import GenerationConfig


MODEL_PATH = '/scratch/gpfs/ll0329/hf_saved_object/chat-llama2-13b'
TOKENIZER_PATH = '/scratch/gpfs/ll0329/hf_saved_object/chat-llama2-13b-tokenizer'
# MODEL_PATH = '/scratch/gpfs/ll0329/hf_saved_object/meta-llama/Meta-Llama-3.1-8B-Instruct'
# TOKENIZER_PATH = '/scratch/gpfs/ll0329/hf_saved_object/meta-llama/Meta-Llama-3.1-8B-Instruct-tokenizer'
PROMPT_PATH = '/home/ll0329/llm_interpretability_persona/marketing_prompts/prompts_toothpaste.txt'
OUTPUT_DIRECTORY = 'chat_llama2_13b_0905'
# NUMBER_OF_ATT_BLOCKS = 32 # 13b has 40 attention blocks, 32 is for 8b llama 3
NUMBER_OF_ATT_BLOCKS = 40 # 13b has 40 attention blocks, 32 is for 8b llama 3
NUMBER_OF_TOP_IDS = 30 # top 30 words from tuned lens
TOKENIZER_SPACE_ID = 259
MAX_SEQ_LEN = 1024

# Configure model parallelism
#torch.cuda.set_device(0)  # Set the primary CUDA device
#torch.cuda.set_device('cpu')  # Set the secondary CUDA device
device_map = "auto"  # Let the library automatically distribute the model across available GPUs

# Load the model
model = AutoModelForCausalLM.from_pretrained(
    MODEL_PATH,
    device_map=device_map,
    torch_dtype=torch.float16,  # Use half-precision to reduce memory usage
    low_cpu_mem_usage=True

)
print(model)

prompts = []
with open(PROMPT_PATH, 'r') as file:
    for line in file:
        # Add each line to the list, stripping newline characters
        prompts.append(line.strip())

print(prompts)


tokenizer = AutoTokenizer.from_pretrained(TOKENIZER_PATH)

temp_dir = TemporaryDirectory()
directory_path = temp_dir.name

tuned_lens = TunedLens.from_model(model)
tuned_lens.save(directory_path)
tuned_lens = TunedLens.from_model_and_pretrained(model, directory_path)
tuned_lens = tuned_lens.to(model.device)
# logit_lens = LogitLens.from_model(model)


for prompt_idx, prompt in enumerate(prompts):
    dest_file = open(f'./{OUTPUT_DIRECTORY}/{prompt_idx}.txt', 'w', encoding='utf-8')
    dest_file.write('PROMPT: ' + prompt + '\n')
    dest_file.write('='*200 + '\n')
    dest_file.flush()

    text = prompt
    input_ids = tokenizer.encode(text)
    targets = input_ids[1:] + [tokenizer.eos_token_id]

    input_ids = torch.tensor(input_ids).to(model.device)
    targets = torch.tensor(targets).to(model.device)
    print ('devices:', model.device, input_ids.device, targets.device)

    pred_traj = PredictionTrajectory.from_lens_and_model(
        lens=tuned_lens,
        model=model,
        input_ids=input_ids,
        tokenizer=tokenizer,
        targets=targets,
    )

    for token_idx, token in enumerate(input_ids):
        for b in range(NUMBER_OF_ATT_BLOCKS):
            top_ids = torch.topk(torch.tensor(pred_traj.log_probs[b, token_idx]), NUMBER_OF_TOP_IDS)
            indices_with_space = [item for elem in top_ids.indices for item in (elem, torch.tensor(TOKENIZER_SPACE_ID))]
            dest_file.write(tokenizer.decode([token]) + ': ' + tokenizer.decode(indices_with_space) + '\n')
            dest_file.flush()

    generation_config = GenerationConfig(
        pad_token_id=tokenizer.eos_token_id,
        eos_token_id=tokenizer.eos_token_id,
        top_k = 10,
        top_p = 0.9,
        no_repeat_ngram_size=3,
        do_sample=True,
        num_return_sequences=1,
        repetition_penalty=1.15,
        max_new_tokens=MAX_SEQ_LEN,
    )
    tokenizer.pad_token_id = tokenizer.eos_token_id
    model_inputs = tokenizer(text, padding='longest', truncation=True, max_length=MAX_SEQ_LEN, return_tensors='pt').to('cuda')
    output = model.generate(**model_inputs, generation_config=generation_config)
    output_text = tokenizer.batch_decode(output, skip_special_tokens=True)
    dest_file.write('*'*200 + '\n')
    dest_file.write('GENERATED TEXT: ' + output_text[0].replace(text, '') + '\n')
    dest_file.flush()
    dest_file.close()
```
